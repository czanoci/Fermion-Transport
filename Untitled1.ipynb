```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import expm
from scipy.optimize import minimize

# 1. Utility functions: NS loadings and AFNS adjustment
def ns_loadings(lmbda, taus):
    """Nelson-Siegel factor loadings B."""
    term = (1 - np.exp(-lmbda * taus)) / (lmbda * taus)
    return np.vstack([np.ones_like(taus), term, term - np.exp(-lmbda * taus)]).T

def afns_C(sigma, lmbda, taus):
    """Yield adjustment term C for AFNS."""
    # following Christensen (2009) formula for diagonal sigma
    A = np.sum(sigma[0]**2)
    B = np.sum(sigma[1]**2)
    C = np.sum(sigma[2]**2)
    D = np.dot(sigma[0], sigma[1])
    E = np.dot(sigma[0], sigma[2])
    F = np.dot(sigma[1], sigma[2])
    t = taus
    la = lmbda
    # precompute exp terms
    e1 = np.exp(-la * t)
    e2 = np.exp(-2 * la * t)
    # formulas
    r1 = -A * t**2 / 6
    r2 = -B * (1/(2*la**2) - (1 - e1)/(la**3 * t) + (1 - e2)/(4*la**3 * t))
    r3 = -C * (1/(2*la**2) + e1/(la**2) - t*e2/(4*la) - 3*e2/(4*la**2)
               - 2*(1-e1)/(la**3 * t) + 5*(1-e2)/(8*la**3 * t))
    r4 = -D * (t/(2*la) + e1/(la**2) - (1-e1)/(la**3 * t))
    r5 = -E * (3*e1/(la**2) + t/(2*la) + t*e1/la - 3*(1-e1)/(la**3 * t))
    r6 = -F * (1/(la**2) + e1/(la**2) - e2/(2*la**2)
               - 3*(1-e1)/(la**3 * t) + 3*(1-e2)/(4*la**3 * t))
    return r1 + r2 + r3 + r4 + r5 + r6

# 2. Simulate synthetic AFNS data
def simulate_afns(params, taus, N, dt=1/252):
    """Simulate X_t factors and yields with measurement noise."""
    # unpack
    kappa = np.diag(params['kappa'])
    sigma = np.diag(params['sigma'])
    theta = params['theta']
    lmbda = params['lambda']
    meas_sd = params['meas_sd']

    # precompute
    B = ns_loadings(lmbda, taus)
    C = afns_C(sigma, lmbda, taus)

    # allocate
    X = np.zeros((N, 3))
    yields = np.zeros((N, len(taus)))
    X_prev = theta.copy()

    # Euler-Maruyama
    for t in range(N):
        dW = np.random.randn(3) * np.sqrt(dt)
        X_prev = X_prev + kappa.dot(theta - X_prev) * dt + sigma.dot(dW)
        X[t] = X_prev
        true_y = B.dot(X_prev) + C
        yields[t] = true_y + np.random.randn(len(taus)) * meas_sd

    return X, yields

# 3. Log-likelihood via Kalman filter
def kalman_loglik(para, yields, taus, dt=1/252):
    # unpack linear params
    kappa = np.diag(para[:3])
    sigma = np.diag(para[3:6])
    theta = para[6:9]
    lmbda = para[9]
    meas_sd = para[10]

    B = ns_loadings(lmbda, taus)
    C = afns_C(sigma, lmbda, taus)
    H = np.eye(len(taus)) * meas_sd**2

    # Discretize
    Phi = expm(-kappa * dt)
    mu = (np.eye(3) - Phi).dot(theta)
    # process noise covariance
    Q = np.zeros((3,3))
    # approximate integral for Q
    for i in range(3):
        for j in range(3):
            Q[i,j] = (sigma[i,i]*sigma[j,j] * (1 - np.exp(-(kappa[i,i]+kappa[j,j])*dt))
                      / (kappa[i,i] + kappa[j,j]))

    # init
    xhat = theta.copy()
    P = np.linalg.solve(kappa + kappa.T, sigma.dot(sigma.T))  # stationary cov

    ll = 0
    T, M = yields.shape

    for t in range(T):
        # predict
        x_pred = mu + Phi.dot(xhat)
        P_pred = Phi.dot(P).dot(Phi.T) + Q
        # observe
        y = yields[t]
        v = y - B.dot(x_pred) - C
        S = B.dot(P_pred).dot(B.T) + H
        # gain
        K = P_pred.dot(B.T).dot(np.linalg.inv(S))
        # update
        xhat = x_pred + K.dot(v)
        P = P_pred - K.dot(B).dot(P_pred)
        # accumulate log-lik
        ll += -0.5 * (M*np.log(2*np.pi) + np.log(np.linalg.det(S)) + v.dot(np.linalg.solve(S, v)))

    return -ll  # negative for minimization

# 4. Set true parameters and generate data
true_params = {
    'kappa': np.array([0.5, 0.3, 0.2]),
    'sigma': np.array([0.02, 0.015, 0.01]),
    'theta': np.array([0.03, -0.01, 0.01]),
    'lambda': 0.7,
    'meas_sd': 0.002
}
taus = np.array([0.5, 1, 2, 5, 10])
N = 300

X_true, yields = simulate_afns(true_params, taus, N)

# 5. Optimize
init = np.array([0.4, 0.2, 0.1, 0.015, 0.01, 0.005, 0.02, -0.005, 0.005, 0.5, 0.001])
bounds = [(1e-3, 5)]*3 + [(1e-4, 1)]*3 + [(-1,1)]*3 + [(1e-2, 2)] + [(1e-4,0.01)]
res = minimize(kalman_loglik, init, args=(yields, taus), bounds=bounds, method='L-BFGS-B')
est = res.x

# 6. Display results
param_names = ['kappa1','kappa2','kappa3',
               'sigma1','sigma2','sigma3',
               'theta1','theta2','theta3','lambda','meas_sd']
print("True vs Estimated Parameters:")
for name, true_v, est_v in zip(param_names, list(true_params['kappa'])+list(true_params['sigma'])+
                                list(true_params['theta'])+[true_params['lambda'], true_params['meas_sd']], est):
    print(f"{name:8s}  {true_v:8.4f}  {est_v:8.4f}")

# 7. Plot last yield curve
B_est = ns_loadings(est[9], taus)
C_est = afns_C(np.diag(est[3:6]), est[9], taus)
X_est, _ = res.fun, None  # factor estimates not stored here

plt.figure()
plt.plot(taus, yields[-1]*100, 'o-', label='Observed')
plt.plot(taus, B_est.dot(X_true[-1]) + C_est*100, 'x--', label='True Model Fit')
plt.xlabel('Maturity (Years)')
plt.ylabel('Yield (%)')
plt.title('Final Yield Curve (Synthetic)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Affine Nelson-Siegel (AFNS) model with Kalman filtering, smoothing, and EM calibration
import numpy as np
from scipy.optimize import minimize
from scipy.linalg import expm
import matplotlib.pyplot as plt

# Nelson-Siegel factor loading

def NS_B(lambda_, tau):
    col1 = np.ones_like(tau)
    col2 = (1 - np.exp(-lambda_ * tau)) / (lambda_ * tau)
    col3 = col2 - np.exp(-lambda_ * tau)
    return np.column_stack((col1, col2, col3))

def AFNS_C(sigma, lambda_, tau):
    s = sigma
    A = s[0, 0] ** 2 + s[0, 1] ** 2 + s[0, 2] ** 2
    B = s[1, 0] ** 2 + s[1, 1] ** 2 + s[1, 2] ** 2
    C = s[2, 0] ** 2 + s[2, 1] ** 2 + s[2, 2] ** 2
    D = s[0, 0] * s[1, 0] + s[0, 1] * s[1, 1] + s[0, 2] * s[1, 2]
    E = s[0, 0] * s[2, 0] + s[0, 1] * s[2, 1] + s[0, 2] * s[2, 2]
    F = s[1, 0] * s[2, 0] + s[1, 1] * s[2, 1] + s[1, 2] * s[2, 2]
    t = tau
    la = lambda_
    r1 = -A * t ** 2 / 6
    r2 = -B * (1 / (2 * la ** 2) - (1 - np.exp(-la * t)) / (la ** 3 * t) + (1 - np.exp(-2 * la * t)) / (4 * la ** 3 * t))
    r3 = -C * (1 / (2 * la ** 2) + np.exp(-la * t) / (la ** 2) - t * np.exp(-2 * la * t) / (4 * la) -
               3 * np.exp(-2 * la * t) / (4 * la ** 2) - 2 * (1 - np.exp(-la * t)) / (la ** 3 * t) +
               5 * (1 - np.exp(-2 * la * t)) / (8 * la ** 3 * t))
    r4 = -D * (t / (2 * la) + np.exp(-la * t) / (la ** 2) - (1 - np.exp(-la * t)) / (la ** 3 * t))
    r5 = -E * (3 * np.exp(-la * t) / (la ** 2) + t / (2 * la) + t * np.exp(-la * t) / la -
               3 * (1 - np.exp(-la * t)) / (la ** 3 * t))
    r6 = -F * (1 / (la ** 2) + np.exp(-la * t) / (la ** 2) - np.exp(-2 * la * t) / (2 * la ** 2) -
               3 * (1 - np.exp(-la * t)) / (la ** 3 * t) + 3 * (1 - np.exp(-2 * la * t)) / (4 * la ** 3 * t))
    return r1 + r2 + r3 + r4 + r5 + r6

# Kalman filter and smoother for AFNS

def kalman_afns(yields, v_mat, params, dt):
    nf = 3
    kappa = np.diag(params[:3])
    sigma = np.diag(params[3:6])
    theta = params[6:9]
    lambda_ = params[9]
    meas_sd = np.clip(params[10], 1e-6, None)
    H = np.eye(len(v_mat)) * meas_sd**2

    B = NS_B(lambda_, v_mat)
    C = AFNS_C(sigma, lambda_, v_mat)

    Phi1 = expm(-kappa * dt)
    Q = sigma @ sigma.T * dt
    mu = (np.eye(nf) - Phi1) @ theta

    n_obs = yields.shape[0]
    xhat = theta.copy()
    P = np.eye(nf) * 0.01

    x_preds, P_preds, x_filts, P_filts = [], [], [], []
    for t in range(n_obs):
        x_pred = mu + Phi1 @ xhat
        P_pred = Phi1 @ P @ Phi1.T + Q
        y_hat = B @ x_pred + C
        S = B @ P_pred @ B.T + H
        err = yields[t] - y_hat
        K = P_pred @ B.T @ np.linalg.inv(S)
        xhat = x_pred + K @ err
        P = P_pred - K @ B @ P_pred
        x_preds.append(x_pred)
        P_preds.append(P_pred)
        x_filts.append(xhat.copy())
        P_filts.append(P.copy())

    x_smooth = x_filts.copy()
    for t in reversed(range(n_obs - 1)):
        Ck = P_filts[t] @ Phi1.T @ np.linalg.inv(P_preds[t+1])
        x_smooth[t] = x_filts[t] + Ck @ (x_smooth[t+1] - x_preds[t+1])

    return np.array(x_filts), np.array(x_smooth)

# Log-likelihood for AFNS

def neg_log_likelihood_afns(params, yields, v_mat, dt):
    x_filt, _ = kalman_afns(yields, v_mat, params, dt)
    lambda_ = params[9]
    sigma = np.diag(params[3:6])
    B = NS_B(lambda_, v_mat)
    C = AFNS_C(sigma, lambda_, v_mat)
    meas_sd = np.clip(params[10], 1e-6, None)
    H = np.eye(len(v_mat)) * meas_sd**2

    fitted = B @ x_filt.T + C[:, None]
    err = yields.T - fitted
    ll = -0.5 * np.sum((err.T @ np.linalg.inv(H)) * err.T) - 0.5 * yields.shape[0] * np.log(np.linalg.det(H))
    return -ll

# EM algorithm for AFNS

def em_afns(yields, v_mat, dt, max_iter=10):
    init_params = np.array([
        0.2, 0.15, 0.1,  # kappa
        0.01, 0.015, 0.02,  # sigma
        0.02, -0.01, 0.005,  # theta
        0.5,  # lambda
        0.001  # meas_sd
    ])

    for _ in range(max_iter):
        def objective(p):
            return neg_log_likelihood_afns(p, yields, v_mat, dt)

        res = minimize(objective, init_params, method='L-BFGS-B', options={'disp': False, 'maxiter': 200})
        if res.success:
            init_params = res.x
        else:
            break

    return res

# Example use with synthetic data
if __name__ == "__main__":
    v_mat = np.array([0.5, 1, 2, 5, 10])
    dt = 1/52
    np.random.seed(42)
    true_params = np.array([
        0.2, 0.15, 0.1,
        0.01, 0.015, 0.02,
        0.02, -0.01, 0.005,
        0.5,
        0.001
    ])
    nf = 3
    x = np.zeros((260, nf))
    for t in range(1, 260):
        x[t] = x[t-1] + (-np.diag(true_params[:3]) @ (x[t-1] - true_params[6:9]) * dt + np.diag(true_params[3:6]) @ np.random.randn(nf) * np.sqrt(dt))
    B = NS_B(true_params[9], v_mat)
    C = AFNS_C(np.diag(true_params[3:6]), true_params[9], v_mat)
    y = B @ x.T + C[:, None] + np.random.normal(0, 0.001, size=(len(v_mat), 260))

    res = em_afns(y.T, v_mat, dt)

    x_filt, x_smooth = kalman_afns(y.T, v_mat, res.x, dt)
    for i in range(nf):
        plt.plot(x_filt[:, i], label=f"Filt {i+1}")
        plt.plot(x_smooth[:, i], '--', label=f"Smooth {i+1}")
    plt.legend()
    plt.title("AFNS Filtered vs Smoothed Factors")
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
from scipy.linalg import expm

# Nelson-Siegel loadings

def ns_loadings(lambda_, tau):
    tau = np.atleast_1d(tau)
    l_tau = lambda_ * tau
    with np.errstate(divide='ignore', invalid='ignore'):
        col2 = np.where(tau == 0, 1.0, (1 - np.exp(-l_tau)) / l_tau)
        col3 = col2 - np.exp(-l_tau)
    return np.column_stack([np.ones_like(tau), col2, col3])

# Yield adjustment term C(tau)
def afns_C(sigma, lambda_, tau):
    A = sigma @ sigma.T
    C_vals = []
    for t in tau:
        B_vals = np.zeros((3, 3))
        for i in range(3):
            for j in range(3):
                den = lambda_ * t * (i + 1) * (j + 1)
                num = 1 - np.exp(-lambda_ * t)
                B_vals[i, j] = num / den
        C_vals.append(-0.5 * np.trace(A @ B_vals))
    return np.array(C_vals)

# Simulate AFNS process under P

def simulate_afns(kappa_p, theta_p, sigma, lambda_, tau, dt, n_obs):
    nf = 3
    X = np.zeros((n_obs, nf))
    for t in range(1, n_obs):
        dX = -kappa_p @ (X[t-1] - theta_p) * dt + sigma @ np.random.randn(nf) * np.sqrt(dt)
        X[t] = X[t-1] + dX
    B = ns_loadings(lambda_, tau)
    C = afns_C(sigma, lambda_, tau)
    yields = X @ B.T + C + 0.002 * np.random.randn(n_obs, len(tau))
    return yields, X

# Step 1: Estimate Q dynamics using Kalman filter with measurement error
def step1_estimate_q(yields, tau, dt, init_params):
    def neg_log_likelihood_q(params):
        kappa_diag = np.diag(params[:3])
        sigma_diag = np.diag(params[3:6])
        theta = params[6:9]
        lambda_ = params[9]
        meas_sd = np.clip(params[10:], 1e-6, None)
        H = np.diag(meas_sd ** 2)

        B = ns_loadings(lambda_, tau)
        C = afns_C(sigma_diag, lambda_, tau)

        n_obs, n_tau = yields.shape
        nf = 3
        X = np.zeros((n_obs, nf))
        xhat = theta.copy()
        P = np.eye(nf) * 0.01
        Q = sigma_diag @ sigma_diag.T * dt
        Phi = expm(-kappa_diag * dt)
        mu = (np.eye(nf) - Phi) @ theta
        ll = 0.0

        for t in range(n_obs):
            # Prediction
            x_pred = mu + Phi @ xhat
            P_pred = Phi @ P @ Phi.T + Q

            y_hat = B @ x_pred + C
            S = B @ P_pred @ B.T + H
            err = yields[t] - y_hat

            K = P_pred @ B.T @ np.linalg.inv(S)
            xhat = x_pred + K @ err
            P = P_pred - K @ B @ P_pred

            X[t] = xhat
            ll += -0.5 * (np.log(np.linalg.det(S)) + err @ np.linalg.inv(S) @ err + n_tau * np.log(2 * np.pi))

        return -ll

    result = minimize(neg_log_likelihood_q, init_params, method='L-BFGS-B')
    return result.x

# Step 2: Estimate P dynamics from Kalman-smoothed factors

def step2_estimate_p(X, dt):
    dX = (X[1:] - X[:-1]) / dt
    X_mid = (X[1:] + X[:-1]) / 2
    beta = np.linalg.lstsq(X_mid, dX, rcond=None)[0].T
    kappa_p = -beta
    theta_p = -np.linalg.inv(kappa_p) @ np.mean(dX - beta @ X_mid.T, axis=1)
    return kappa_p, theta_p

# Iterative calibration procedure with diagnostics

def calibrate_afns_iterative(yields, tau, dt, n_iter=5, plot=True):
    nf = 3
    n_tau = len(tau)
    init_params = np.array([
        0.1, 0.1, 0.1,       # kappa
        0.01, 0.01, 0.01,    # sigma
        0.01, -0.01, 0.0,    # theta
        0.5,                # lambda
    ] + [0.002] * n_tau)     # measurement errors

    for i in range(n_iter):
        # Step 1: Estimate Q dynamics
        params_q = step1_estimate_q(yields, tau, dt, init_params)

        # Simulate Kalman-smoothed factors using estimated Q dynamics
        kappa_q = np.diag(params_q[:3])
        sigma_q = np.diag(params_q[3:6])
        theta_q = params_q[6:9]
        lambda_ = params_q[9]
        meas_sd = np.clip(params_q[10:], 1e-6, None)
        H = np.diag(meas_sd ** 2)

        B = ns_loadings(lambda_, tau)
        C = afns_C(sigma_q, lambda_, tau)

        n_obs = yields.shape[0]
        X = np.zeros((n_obs, nf))
        xhat = theta_q.copy()
        P = np.eye(nf) * 0.01
        Q = sigma_q @ sigma_q.T * dt
        Phi = expm(-kappa_q * dt)
        mu = (np.eye(nf) - Phi) @ theta_q

        for t in range(n_obs):
            x_pred = mu + Phi @ xhat
            P_pred = Phi @ P @ Phi.T + Q
            y_hat = B @ x_pred + C
            S = B @ P_pred @ B.T + H
            err = yields[t] - y_hat

            K = P_pred @ B.T @ np.linalg.inv(S)
            xhat = x_pred + K @ err
            P = P_pred - K @ B @ P_pred
            X[t] = xhat

        # Step 2: Estimate P dynamics
        kappa_p, theta_p = step2_estimate_p(X, dt)

        # Update initial params for next Q estimation
        init_params[:3] = np.diag(kappa_p)
        init_params[6:9] = theta_p

    if plot:
        # Plot estimated factors
        plt.figure(figsize=(12, 4))
        for i in range(nf):
            plt.plot(X[:, i], label=f'Factor {i+1}')
        plt.title('Estimated Latent Factors')
        plt.xlabel('Time')
        plt.legend()
        plt.grid(True)
        plt.show()

        # Plot fitted vs actual yields
        fitted_yields = X @ B.T + C
        plt.figure(figsize=(12, 6))
        for i in range(min(5, n_tau)):
            plt.plot(yields[:, i], label=f'Observed ={tau[i]}')
            plt.plot(fitted_yields[:, i], '--', label=f'Fitted ={tau[i]}')
        plt.title('Observed vs Fitted Yields')
        plt.xlabel('Time')
        plt.ylabel('Yield')
        plt.legend()
        plt.grid(True)
        plt.show()

    return params_q, kappa_p, theta_p, X
```

```python
# Rebonato-style 3-factor Affine Term Structure Model (ATSM) MLE Calibration
import numpy as np
from scipy.optimize import minimize
from scipy.linalg import expm, solve_continuous_lyapunov
import matplotlib.pyplot as plt

# Define yield function from ATSM with exponential affine form
def affine_yields(X, A, B):
    return A + X @ B.T

# Simulate latent factors X_t under P measure
def simulate_factors(kappa, theta, sigma, dt, n_obs):
    nf = kappa.shape[0]
    X = np.zeros((n_obs, nf))
    for t in range(1, n_obs):
        dX = -kappa @ (X[t-1] - theta) * dt + sigma @ np.random.randn(nf) * np.sqrt(dt)
        X[t] = X[t-1] + dX
    return X

# Solve Riccati ODEs to get A(tau) and B(tau) for affine yields
def build_A_B(kappa, sigma, rho, tau):
    nf = kappa.shape[0]
    A = np.zeros(len(tau))
    B = np.zeros((len(tau), nf))
    for i, t in enumerate(tau):
        exp_neg_kappa_t = expm(-kappa * t)
        B[i, :] = np.linalg.solve(kappa, (np.eye(nf) - exp_neg_kappa_t)) @ np.ones(nf)
        Sigma = sigma @ rho @ sigma.T
        phi = solve_continuous_lyapunov(kappa.T, -Sigma)
        A[i] = -0.5 * np.trace(phi @ B[i, :][:, None] @ B[i, :][None, :])
    return A, B

# Kalman filter with storage for smoothing
def kalman_filter_and_smoother(yields, A, B, kappa, theta, sigma, H, dt):
    nf = len(theta)
    n_obs = yields.shape[0]
    X = np.zeros((n_obs, nf))
    P = np.eye(nf) * 0.01
    Phi = expm(-kappa * dt)
    Q = sigma @ sigma.T * dt
    mu = (np.eye(nf) - Phi) @ theta

    xhat = theta.copy()
    x_preds = []
    P_preds = []
    x_filts = []
    P_filts = []

    for t in range(n_obs):
        x_pred = mu + Phi @ xhat
        P_pred = Phi @ P @ Phi.T + Q

        y_hat = A + B @ x_pred
        S = B @ P_pred @ B.T + H
        err = yields[t] - y_hat

        K = P_pred @ B.T @ np.linalg.inv(S)
        xhat = x_pred + K @ err
        P = P_pred - K @ B @ P_pred

        x_preds.append(x_pred)
        P_preds.append(P_pred)
        x_filts.append(xhat.copy())
        P_filts.append(P.copy())

    x_smooth = x_filts.copy()
    for t in reversed(range(n_obs - 1)):
        C = P_filts[t] @ Phi.T @ np.linalg.inv(P_preds[t+1])
        x_smooth[t] = x_filts[t] + C @ (x_smooth[t+1] - x_preds[t+1])

    return np.array(x_filts), np.array(x_smooth)

# Log-likelihood function
def neg_log_likelihood(params, yields, tau, dt):
    nf = 3
    expected_len = nf**2 + nf + nf**2 + 1
    if len(params) != expected_len:
        raise ValueError(f"Expected {expected_len} parameters but got {len(params)}")

    kappa = params[:nf**2].reshape(nf, nf)
    theta = params[nf**2:nf**2+nf]
    sigma_flat = params[nf**2+nf:nf**2+nf+nf**2]
    sigma = sigma_flat.reshape(nf, nf)
    meas_sd = np.clip(params[-1], 1e-6, None)
    H = np.eye(len(tau)) * meas_sd**2

    rho = np.eye(nf)
    A, B = build_A_B(kappa, sigma, rho, tau)

    X_filt, _ = kalman_filter_and_smoother(yields, A, B, kappa, theta, sigma, H, dt)
    fitted_yields = affine_yields(X_filt, A, B)
    err = yields - fitted_yields
    ll = -0.5 * np.sum((err @ np.linalg.inv(H) * err).sum(axis=1) + np.log(np.linalg.det(H)) + len(tau) * np.log(2 * np.pi))

    return -ll

# Wrapper to calibrate model
def calibrate_rebonato(yields, tau, dt):
    nf = 3
    n_params = nf**2 + nf + nf**2 + 1
    init_params = np.random.normal(0, 0.1, size=n_params)

    result = minimize(
        neg_log_likelihood,
        init_params,
        args=(yields, tau, dt),
        method='L-BFGS-B',
        options={'disp': True, 'maxiter': 1000}
    )

    return result

# Example usage with synthetic data
if __name__ == "__main__":
    tau = np.array([0.5, 1.0, 2.0, 5.0, 10.0])
    dt = 1/52
    true_kappa = np.array([[0.2, 0.01, 0.0], [0.0, 0.1, 0.02], [0.01, 0.0, 0.05]])
    true_theta = np.array([0.01, -0.005, 0.002])
    true_sigma = np.array([[0.01, 0.002, 0.001], [0.0, 0.015, 0.003], [0.0, 0.0, 0.02]])

    X_true = simulate_factors(true_kappa, true_theta, true_sigma, dt, 260)
    A_true, B_true = build_A_B(true_kappa, true_sigma, np.eye(3), tau)
    y_synth = affine_yields(X_true, A_true, B_true) + np.random.normal(0, 0.001, size=(260, len(tau)))

    res = calibrate_rebonato(y_synth, tau, dt)

    if res.success:
        print("MLE succeeded.")
        est_params = res.x
        print(est_params)
    else:
        print("MLE failed:", res.message)
```
